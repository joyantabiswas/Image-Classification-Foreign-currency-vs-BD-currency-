```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from skimage.color import rgb2gray
import cv2 #opencv library
from skimage.filters import (threshold_otsu, threshold_niblack,threshold_sauvola)
from skimage.util import img_as_ubyte
from skimage import io
from skimage.morphology import erosion, dilation, opening, closing
from skimage.morphology import disk
```

```python
def image_to_feature_vector(images, size=(32, 32)):
    '''This function Resizes the image to a fixed size, then 
      flatten the image into a list of raw pixel intensities'''

    return cv2.resize(images, size).flatten()
```

```python

def read_data_set(folder , is_labelled):
    '''Reads the Image Dataset'''
    rawImages = []
    labels = []
    # reads the images one by one and converts them into raw pixed feature vector
    for filename in os.listdir(folder):
        image = cv2.imread(os.path.join(folder, filename))
        grayscale_image=rgb2gray(image)
        window_size = 25
        phantom_picture_bin = img_as_ubyte( grayscale_image)
        selem=disk(6)
        label = filename.split("_")[0]  # 'BDT' or 'FC' from the filename
        labels.append(label)
        # extract raw pixel intensity "features", followed by a color histogram 
        # to characterize the color distribution of the pixels in the image

        if value == 1: 
           pixel_values = image_to_feature_vector(image)
           #plt.imshow(image)
           #plt.show()
                                                   #*********************GRAYSCALE_PART********************
        elif (value == 2):
           gray_img_bin= img_as_ubyte(grayscale_image)
           pixel_values = image_to_feature_vector(gray_img_bin)
           #plt.imshow(grayscale_image, cmap=plt.cm.gray)
           #plt.show()
                                                    #*********************SEGEMENTATION_PART********************
                           
                                                                        #OTSU_SEGEMENTATION
        elif (value==3):
           binary_global =grayscale_image > threshold_otsu(grayscale_image)
           otsu_img_bin=  img_as_ubyte(binary_global)
           pixel_values= image_to_feature_vector(otsu_img_bin) 
           #plt.imshow(binary_global, cmap=plt.cm.gray)
           #plt.show()

                                                                       #NIBLACK_SEGEMENTATION
        
        elif (value == 4):
           thresh_niblack = threshold_niblack(grayscale_image, window_size=window_size, k=0.8)
           binary_niblack = grayscale_image> thresh_niblack
           niblack_img_bin=  img_as_ubyte(binary_niblack)
           pixel_values = image_to_feature_vector(niblack_img_bin)  
           #plt.imshow(binary_niblack, cmap=plt.cm.gray)
           #plt.show()
                      
                                                                      #SAUVOLA_SEGEMENTATION
        elif (value == 5):
           thresh_sauvola = threshold_sauvola(grayscale_image, window_size=window_size)
           binary_sauvola = grayscale_image > thresh_sauvola 
           sauvola_img_bin=  img_as_ubyte(binary_sauvola) 
           pixel_values = image_to_feature_vector(sauvola_img_bin)
           #plt.imshow(binary_sauvola, cmap=plt.cm.gray)
           #plt.show()
            
                                                    #*********************MORPHOLOGICAL_PART*************************
                                                                      
                                                                      #EROSION_PART
        elif (value == 6):
           eroded_picture = erosion(phantom_picture_bin, selem) 
           eroded_picture_bin =  img_as_ubyte(eroded_picture)
           pixel_values = image_to_feature_vector(eroded_picture_bin)
           #plt.imshow(eroded_picture, cmap=plt.cm.gray)
           #plt.show()
                                                                      #DILATION_PART
        elif (value == 7):
           dilated_picture = dilation(phantom_picture_bin, selem)
           dilated_picture_bin =  img_as_ubyte( dilated_picture)
           pixel_values = image_to_feature_vector(dilated_picture_bin)
           #plt.imshow(dilated_picture, cmap=plt.cm.gray)
           #plt.show()
                                                                      #OPENING_PART
        
        elif (value == 8):
           opened_picture = opening(phantom_picture_bin, selem)
           opened_picture_bin = img_as_ubyte( opened_picture)
           pixel_values = image_to_feature_vector(opened_picture_bin) 
           #plt.imshow(opened_picture, cmap=plt.cm.gray)
           #plt.show()
                                                                      #CLOSING_PART
        elif (value == 9):
           closed_picture = closing(phantom_picture_bin, selem)
           closed_picture_bin =  img_as_ubyte( closed_picture)
           pixel_values = image_to_feature_vector(closed_picture_bin) 
           #plt.imshow(closed_picture, cmap=plt.cm.gray)
           #plt.show()
           
        elif (value == 10):
           dilated_picture = dilation(phantom_picture_bin, selem)
           erosion_on_dilated_picture=erosion(dilated_picture, selem)
           erosion_on_dilated_picture_bin = img_as_ubyte(erosion_on_dilated_picture)
           pixel_values = image_to_feature_vector(erosion_on_dilated_picture_bin)
           #plt.imshow(erosion_on_dilated_picture, cmap=plt.cm.gray)
           #plt.show() 
        elif (value == 11):
           opened_picture = opening(phantom_picture_bin, selem)
           erosion_on_opened_picture=erosion(opened_picture, selem)
           erosion_on_opened_picture_bin = img_as_ubyte(erosion_on_opened_picture)
           pixel_values = image_to_feature_vector(erosion_on_opened_picture_bin) 
           #plt.imshow( erosion_on_opened_picture, cmap=plt.cm.gray)
           #plt.show()
        elif (value == 12):
           closed_picture = closing(phantom_picture_bin, selem)
           erosion_on_closed_picture=erosion(closed_picture, selem)
           erosion_on_closed_picture_bin = img_as_ubyte(erosion_on_closed_picture)
           pixel_values = image_to_feature_vector(erosion_on_closed_picture_bin) 
           #plt.imshow(erosion_on_closed_picture, cmap=plt.cm.gray)
           #plt.show()
        elif (value == 13):
           eroded_picture = erosion(phantom_picture_bin, selem) 
           dilation_on_eroded_picture=dilation(eroded_picture, selem)
           dilation_on_eroded_picture_bin = img_as_ubyte(dilation_on_eroded_picture)
           pixel_values = image_to_feature_vector(dilation_on_eroded_picture_bin) 
           #plt.imshow(dilation_on_eroded_picture, cmap=plt.cm.gray)
           #plt.show()
        elif (value == 14):
           opened_picture = opening(phantom_picture_bin, selem)
           dilation_on_opened_picture=dilation(opened_picture, selem)
           dilation_on_opened_picture_bin = img_as_ubyte(dilation_on_opened_picture)
           pixel_values = image_to_feature_vector(dilation_on_opened_picture_bin) 
           #plt.imshow(dilation_on_opened_picture, cmap=plt.cm.gray)
           #plt.show()
        elif (value == 15):
           closed_picture = closing(phantom_picture_bin, selem)
           dilation_on_closed_picture=dilation(closed_picture, selem)
           dilation_on_closed_picture_bin = img_as_ubyte(dilation_on_closed_picture)
           pixel_values = image_to_feature_vector(dilation_on_closed_picture_bin) 
           #plt.imshow(dilation_on_closed_picture, cmap=plt.cm.gray)
           #plt.show()
        elif (value == 16):
           eroded_picture = erosion(phantom_picture_bin, selem) 
           opening_on_eroded_picture=opening(eroded_picture, selem)
           opening_on_eroded_picture_bin = img_as_ubyte(opening_on_eroded_picture)
           pixel_values = image_to_feature_vector(opening_on_eroded_picture_bin) 
           #plt.imshow(opening_on_eroded_picture, cmap=plt.cm.gray)
           #plt.show()
        elif (value == 17):
           dilated_picture = dilation(phantom_picture_bin, selem)
           opening_on_dilated_picture=opening(dilated_picture, selem)
           opening_on_dilated_picture_bin = img_as_ubyte(opening_on_dilated_picture)
           pixel_values = image_to_feature_vector(opening_on_dilated_picture_bin)
           #plt.imshow(opening_on_dilated_picture, cmap=plt.cm.gray)
           #plt.show() 
        elif (value == 18):
           closed_picture = closing(phantom_picture_bin, selem)
           opening_on_closed_picture=dilation(closed_picture, selem)
           opening_on_closed_picture_bin = img_as_ubyte(opening_on_closed_picture)
           pixel_values = image_to_feature_vector(opening_on_closed_picture_bin) 
           #plt.imshow(opening_on_closed_picture, cmap=plt.cm.gray)
           #plt.show()
        elif (value == 19):
           eroded_picture = erosion(phantom_picture_bin, selem) 
           closing_on_eroded_picture=opening(eroded_picture, selem)
           closing_on_eroded_picture_bin = img_as_ubyte(closing_on_eroded_picture)
           pixel_values = image_to_feature_vector(closing_on_eroded_picture_bin)
           #plt.imshow(closing_on_eroded_picture, cmap=plt.cm.gray)
           #plt.show()
        elif (value == 20):
           dilated_picture = dilation(phantom_picture_bin, selem)
           closing_on_dilated_picture=opening(dilated_picture, selem)
           closing_on_dilated_picture_bin = img_as_ubyte(closing_on_dilated_picture)
           pixel_values = image_to_feature_vector(closing_on_dilated_picture_bin)
           #plt.imshow(closing_on_dilated_picture, cmap=plt.cm.gray)
           #plt.show()
        elif (value == 21):
           opened_picture = opening(phantom_picture_bin, selem)
           closing_on_opened_picture=dilation(opened_picture, selem)
           closing_on_opened_picture_bin = img_as_ubyte(closing_on_opened_picture)
           pixel_values = image_to_feature_vector(closing_on_opened_picture_bin) 
           #plt.imshow( closing_on_opened_picture, cmap=plt.cm.gray)
           #plt.show()   
        else:
           print("*******************ERROR!********************\nYou have put a wrong integer.")

        rawImages.append(pixel_values)  # features of BDT_1.jpg, FC_1.jpg
       #print(rawImages)
    return rawImages, labels
```
